## Drawing a Bar Plot
To draw a bar plot, you can use the bar function:

````python
majors = ['Computer Science', 'Physics', 'Mathematics', 'Biology']
student_counts = [100, 80, 90, 70]

plt.bar(majors, student_counts)
plt.show()
````

# Changing Bar Colors in Matplotlib

In a bar plot, you can change the color of the bars. This can be done using the `color` parameter in the `bar` function. For example:

````python
majors = ['Computer Science', 'Physics', 'Mathematics', 'Biology']
student_counts = [100, 80, 90, 70]
colors = ['blue', 'green', 'red', 'purple']

plt.bar(majors, student_counts, color=colors)
plt.show()
````

# Advanced Matplotlib Tutorial

Matplotlib provides the capability to create multiple plots in one figure, known as subplots. This can be done using the `subplot` function.

## Creating Multiple Plots in One Figure

To create multiple plots in one figure, you can use the `subplot` function. For example:

```python

````python
import numpy as np

# Create data
x = np.linspace(0, 2 * np.pi, 400)
y1 = np.sin(x)
y2 = np.cos(x)

# Create a figure with two subplots
fig, axs = plt.subplots(2)
fig.suptitle('Vertically stacked subplots')
axs[0].plot(x, y1)
axs[1].plot(x, y2)

plt.show()
````
